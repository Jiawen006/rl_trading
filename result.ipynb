# Visualization of the example trading result

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

file_name = "results/example.csv"
data = pd.read_csv(file_name, header=None)
value = data.to_numpy()
day = np.arange(value.size)

plt.figure(figsize=(8, 6))

# Create a plot
plt.plot(day, value, linewidth=1)

# Label the axes and set a title
plt.xlabel("Trading Day")
plt.ylabel("Asset Value")
plt.title("Visualization of trading dataset performance")

# Display the plot
plt.show()
```
